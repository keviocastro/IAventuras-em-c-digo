```python
import datetime
import random
import pytz

tz = pytz.timezone('America/Sao_Paulo')

def gerar_data_horario_aleatorio(data):
    hora = random.randint(6, 22)
    minuto = random.randint(0, 59)
    segundo = random.randint(0, 59)

    dt = datetime.datetime.strptime(data, "%Y-%m-%d")
    dt = dt.replace(hour=hora, minute=minuto, second=segundo)

    dt = tz.localize(dt)

    return dt.strftime("%Y-%m-%d %H:%M:%S%z")

df['data_checkin'] = df['data_checkin'].apply(gerar_data_horario_aleatorio)
```

```python
df['data_checkin'] = pd.to_datetime(df['data_checkin'], format="%Y-%m-%d %H:%M:%S%z")

def gerar_data_checkout(data_checkin):
    minutos_extra = random.randint(40, 60)
    return data_checkin + datetime.timedelta(minutes=minutos_extra)

df['data_checkout'] = df['data_checkin'].apply(gerar_data_checkout)

df['data_checkout'] = df['data_checkout'].dt.strftime("%Y-%m-%d %H:%M:%S%z")
```
