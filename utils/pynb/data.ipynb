```python
checkins = []

for i, row in df.iterrows():
    nome = row['nome_aluno']
    data_inicio = row['data_inicio']

    n_checkins = np.random.randint(20, 150) # cada aluno terá entre 20 e 150 checkins

    datas_checkin = pd.date_range(start=data_inicio, end='2025-04-03').to_pydatetime()

    if len(datas_checkin) == 0:
        continue

    datas_aleatorias = np.random.choice(datas_checkin, size=n_checkins, replace=True)

    for data in datas_aleatorias:
        checkins.append({
            'nome_aluno': nome,
            'data_checkin': pd.to_datetime(data)
        })

df_checkins = pd.DataFrame(checkins)
df_checkins.sort_values(by=['nome_aluno', 'data_checkin'], inplace=True)
df_checkins.reset_index(drop=True, inplace=True)

df_checkins['id_checkin'] = range(1, len(df_checkins)+1) # adiciona id de checkin
```

```python
import datetime
import random

tz = "America/Sao_Paulo"

def gerar_data_horario_aleatorio(data):
    hora = random.randint(6, 22)
    minuto = random.randint(0, 59)
    segundo = random.randint(0, 59)

    dt = pd.Timestamp(data).replace(hour=hora, minute=minuto, second=segundo)
    return dt.tz_localize(tz)

def gerar_data_checkout(data_checkin):
    minutos_extra = random.randint(40, 60)
    return data_checkin + pd.Timedelta(minutes=minutos_extra)

df_checkins['data_checkin'] = df_checkins['data_checkin'].apply(gerar_data_horario_aleatorio)
df_checkins['data_checkout'] = df_checkins['data_checkin'].apply(gerar_data_checkout)
```

```python
df['data_inicio'] = pd.to_datetime(df['data_inicio'])

def gerar_data_fim(data_inicio):
    if random.random() < 0.3:
        dias_extra = random.randint(30, 730)
        return data_inicio + pd.Timedelta(days=dias_extra)
    return pd.NaT  # Se não cancelar, retorna NaT (indica que o aluno ainda está ativo)

df['data_fim'] = df['data_inicio'].apply(gerar_data_fim)
```

```python
df["data_inicio"] = pd.to_datetime(df['data_inicio'], format="%Y-%m-%d")
```

```python
df_checkins["data_checkout"] = pd.to_datetime(df_checkins['data_checkout'], format="%Y-%m-%d %H:%M:%S%z")
df_checkins["data_checkin"] = pd.to_datetime(df_checkins['data_checkin'], format="%Y-%m-%d %H:%M:%S%z")
```

```python
# Garantir que TODAS as datas sejam UTC e REMOVER o fuso horário
df_checkins["data_checkin"] = df_checkins["data_checkin"].dt.tz_convert("UTC").dt.tz_localize(None)
df_checkins["data_checkout"] = df_checkins["data_checkout"].dt.tz_convert("UTC").dt.tz_localize(None)

df["data_inicio"] = df["data_inicio"].dt.tz_localize("UTC")
df["data_fim"] = df["data_fim"].dt.tz_localize("UTC", ambiguous="NaT")  # Evita erro em NaN
```

```python
# Fazer merge mantendo múltiplos check-ins por aluno
df_merged = df_checkins.merge(df, on="nome_aluno", how="left")
```

```python
df_merged.info()
```

```python
df_merged["data_inicio"] = df_merged["data_inicio"].dt.tz_localize(None)
df_merged["data_fim"] = df_merged["data_fim"].dt.tz_localize(None)
```

```python
df_merged = df_merged[
    (df_merged["data_checkin"] >= df_merged["data_inicio"]) &
    ((df_merged["data_fim"].isna()) | (df_merged["data_checkin"] <= df_merged["data_fim"]))
].copy()
```

```python
df_merged.info()
```

```python
df_merged["data_checkin"] = df_merged["data_checkin"].dt.strftime("%Y-%m-%d %H:%M:%S")
df_merged["data_checkout"] = df_merged["data_checkout"].dt.strftime("%Y-%m-%d %H:%M:%S")
df_merged["data_inicio"] = df_merged["data_inicio"].dt.strftime("%Y-%m-%d %H:%M:%S")
df_merged["data_fim"] = df_merged["data_fim"].dt.strftime("%Y-%m-%d %H:%M:%S")
```

```python
alunos_unicos = df_merged["nome_aluno"].unique()
```

```python
np.random.seed(42)
valores_aleatorios = np.random.uniform(low=69.90, high=229.90, size=len(alunos_unicos))
```

```python
df_planos = pd.DataFrame({"nome_aluno": alunos_unicos, "valor1": valores_aleatorios})
```

```python
df_planos["valor_plano"] = np.select(
    [df_planos["valor1"] <= 159.90, df_planos["valor1"] <= 229.89],
    [69.90, 159.90],
    default=229.90
)
```

```python
df_planos["nome_plano"] = np.select(
    [df_planos["valor_plano"] == 69.90, df_planos["valor_plano"] == 159.90],
    ["Beginner", "Focus"],
    default="Athlete"
)
```

```python
df_planos["tipo_plano"] = np.select(
    [df_planos["valor_plano"] == 69.90, df_planos["valor_plano"] == 159.90],
    ["Mensal", "Semestral"],
    default="Anual"
)
```

```python
df_merged = df_merged.merge(df_planos, on="nome_aluno", how="left")
```

```python
df_merged.columns
```

```python
df_merged["churn"] = df_merged["data_fim"].isna().astype(int)
```

```python
df_merged = df_merged[[
    'id_checkin',
    'nome_aluno',
    'idade_aluno',
    'data_checkin',
    'data_checkout',
    'data_inicio',
    'data_fim',
    'valor_plano',
    'nome_plano',
    'tipo_plano',
    'churn'
]]
```
